```python
plt.figure(figsize=(10, 6))
plt.plot(df['Date'], df['Order_Demand'], marker='o', linestyle='-')
plt.title('Order Demand Over Time')
plt.xlabel('Date')
plt.ylabel('Order Demand')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Apply log1p transformation
import numpy as np
df['log_order_demand'] = np.log1p(df['Order_Demand'])

plt.figure(figsize=(10, 6))
sns.histplot(df['log_order_demand'], bins=30)
plt.title('Log-transformed Order Demand')
plt.show()
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose

df_ts = df.set_index('Date').sort_index()
result = seasonal_decompose(
    df_ts['Order_Demand'].resample('W').mean(), 
    model='additive', 
    period=12
)

result.plot()
plt.xticks(rotation=45)
plt.show()
```

```python
# Choose a real combination with sufficient data
product = df['Product_Code'].unique()[0]       
warehouse = df['Warehouse'].unique()[0]        

# Filter and group
subset = df[(df['Product_Code'] == product) & (df['Warehouse'] == warehouse)]
subset_grouped = subset.groupby('Date')['Order_Demand'].sum().reset_index()

plt.figure(figsize=(12, 6))
plt.plot(subset['Date'], subset['Order_Demand'])

plt.show()
```

##  Data Insights Summary

1. **Dataset Overview:**
   - The dataset contains **169,211** records across **13 columns**, capturing historical retail demand along with store-level and product-level attributes.
   - There are **no missing or duplicate values**, ensuring a clean and ready-to-model dataset.

2. **Product & Store Distribution:**
   - High product diversity observed from `Product_Code`, but a few products dominate the order counts.
   - Demand is spread across multiple warehouses (`Warehouse`), each showing varying levels of activity.

3. **Target Variable (Order_Demand):**
   - The `Order_Demand` variable is **highly skewed**, with most values clustered near the lower end.
   - A **log transformation (`log_order_demand`)** was applied to normalize the distribution and improve modeling performance.

4. **Promotions & Store Status:**
   - The majority of stores were open, and only a small proportion of days had promotions running.
   - Pie charts reveal class imbalance in features like `Promo`, `Open`, and `StateHoliday`.

5. **Time Series Trends:**
   - A clear **increasing trend in order demand** over time is visible.
   - Future modeling can benefit from capturing **seasonality, trend, and holiday effects**.

6. **Feature Distributions:**
   - `Petrol_price` shows a relatively smooth, continuous distribution, which could act as an external regressor for economic influence.
   - Categorical features such as `Product_Category` and `StateHoliday` are well-balanced and can be useful for grouping or filtering data.
